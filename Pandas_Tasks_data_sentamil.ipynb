```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

read the data

```python
filepath ='/adult.csv'
adult_data = pd.read_csv(filepath)
adult_data.head()
```

descriptive statistics for numeric columns

Display the column names of the DataFrame.

```python
print(adult_data.columns)
```

unique values in the 'workclass' column.

```python
adult_data['workclass'].unique()
```

Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
adult_data['income_binary'] = np.where(adult_data['income'] == '<=50K', 0, 1)
```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
adult_data['age_group'] = pd.cut(adult_data['age'], bins=[0, 30, 60, np.inf], labels=['Youth', 'Adult', 'Senior'], right=False)
display(adult_data.head())
```

Rename the 'educational-num' column to 'education_num'.

```python
adult_data.rename(columns={'educational-num': 'education_num'}, inplace=True)
display(adult_data.head())
```

Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.

```python
adult_data['workclass'] = adult_data['workclass'].replace('?', 'Unknown')
adult_data['occupation'] = adult_data['occupation'].replace('?', 'Unknown')
display(adult_data.head())
```

Calculate the average age for each workclass.

```python
average_age_by_workclass = adult_data.groupby('workclass')['age'].mean()
display(average_age_by_workclass)
```

Find the total capital-gain for each gender.

```python
total_capital_gain_by_gender = adult_data.groupby('gender')['capital-gain'].sum()
display(total_capital_gain_by_gender)
```

Get the count of each marital-status category.

```python
marital_status_counts = adult_data['marital-status'].value_counts()
display(marital_status_counts)
```

Find the mean hours-per-week worked by individuals in different education levels.

```python
mean_hours_per_week_by_education = adult_data.groupby('education')['hours-per-week'].mean()
display(mean_hours_per_week_by_education)
```

Calculate the mean, median, and standard deviation of 'fnlwgt' for each race.

```python
fnlwgt_stats_by_race = adult_data.groupby('race')['fnlwgt'].agg(['mean', 'median', 'std'])
display(fnlwgt_stats_by_race)
```

Get the unique values and their counts for the 'native-country' column

```python
native_country_counts = adult_data['native-country'].value_counts()
display(native_country_counts)
```

Filter the dataset to include only individuals with income >50K.

```python
income_greater_than_50k = adult_data[adult_data['income'] == '>50K']
display(income_greater_than_50k.head())
```

Sort the dataset by age in descending order.

```python
adult_data_sorted_by_age = adult_data.sort_values(by='age', ascending=False)
display(adult_data_sorted_by_age.head())
```

Filter the dataset to include only individuals who work more than 40 hours per week.



```python
adult_data_high_hours = adult_data[adult_data['hours-per-week'] > 40]
display(adult_data_high_hours.head())
```

Calculate the percentage of missing values in each column.

```python
missing_values_percentage = adult_data.isnull().sum() / len(adult_data) * 100
display(missing_values_percentage[missing_values_percentage > 0])
```

Calculate the median income by education level and gender.

```python
median_income_by_education_gender = adult_data.groupby(['education', 'gender'])['income_binary'].median()
display(median_income_by_education_gender)
```

Find the top 3 occupations with the highest average capital-gain.

```python
top_3_occupations_capital_gain = adult_data.groupby('occupation')['capital-gain'].mean().nlargest(3)
display(top_3_occupations_capital_gain)
```

Filter the dataset to include individuals who are 'Married-civ-spouse' and have an income >50K.

```python
married_high_income = adult_data[(adult_data['marital-status'] == 'Married-civ-spouse') & (adult_data['income'] == '>50K')]
display(married_high_income.head())
```

Create 5 different charts and write your observations from the chart as comment

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='education_num', data=adult_data, palette='coolwarm')
plt.title('Distribution of Educational Number')
plt.xlabel('Educational Number')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Observations from the plot:
# - The most common educational number is 9, corresponding to 'HS-grad'.
# - Educational numbers 10 ('Some-college') and 13 ('Bachelors') also have high counts.
# - There's a clear decrease in counts as the educational number increases beyond the common levels, suggesting fewer individuals achieve very high education levels.
# - Very low educational numbers (e.g., 1, 2) also have very low counts.
```

### Chart 2: Relationship between 'hours-per-week' and 'income'

This box plot visualizes the distribution of 'hours-per-week' for each 'income' level. This helps to understand if there is a difference in working hours between lower and higher income earners.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='income', y='hours-per-week', data=adult_data, palette='viridis')
plt.title('Hours per Week vs. Income')
plt.xlabel('Income')
plt.ylabel('Hours per Week')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Observations from the plot:
# - Individuals earning '>50K' generally work more hours per week, with a higher median and a wider spread in the upper quartile.
# - The median hours per week for both income groups is around 40 hours, but the upper quartile for '>50K' is significantly higher.
# - There are more outliers working very long hours in the '>50K' income group.
# - The distribution of 'hours-per-week' for those earning '<=50K' is more concentrated around 40 hours, with fewer individuals working very long hours.
```

### Chart 3: Distribution of 'occupation'

This bar plot shows the count of individuals in each 'occupation' category. It helps to understand the prevalence of different job types in the dataset.

```python
plt.figure(figsize=(14, 7))
sns.countplot(x='occupation', data=adult_data, palette='plasma')
plt.title('Distribution of Occupation')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', alpha=0.75)
plt.show()

# Observations from the plot:
# - 'Prof-specialty' and 'Craft-repair' are among the most common occupations.
# - 'Exec-managerial' also represents a significant portion of the workforce.
# - Occupations like 'Armed-Forces' and 'Priv-house-serv' have very low counts.
# - The 'Unknown' category (which replaced '?') also has a notable count, indicating some missing occupation data.
```

### Chart 4: Capital-loss distribution

This histogram visualizes the distribution of the 'capital-loss' column. It helps to understand the frequency and magnitude of capital losses among individuals in the dataset.

```python
plt.figure(figsize=(10, 6))
sns.histplot(adult_data['capital-loss'], kde=True, bins=50, color='red')
plt.title('Distribution of Capital Loss')
plt.xlabel('Capital Loss')
plt.ylabel('Count / Density')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Observations from the plot:
# - A vast majority of individuals have a 'capital-loss' of 0, indicating no capital loss.
# - For individuals with capital loss, the values are heavily skewed towards lower amounts.
# - There are very few instances of high capital loss.
# - This distribution suggests that capital loss is not a common event or is generally small when it occurs.
```

### Chart 5: 'race' vs 'income_binary'

This count plot shows the distribution of 'income_binary' across different 'race' categories. This helps to visualize potential income disparities among different racial groups.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='race', hue='income_binary', data=adult_data, palette='crest')
plt.title('Income Distribution by Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.legend(title='Income', labels=['<=50K', '>50K'])
plt.grid(axis='y', alpha=0.75)
plt.show()

# Observations from the plot:
# - For all racial groups, the count of individuals earning '<=50K' is higher than those earning '>50K'.
# - The 'White' race has the highest overall counts for both income categories.
# - While 'White' and 'Asian-Pac-Islander' groups show a noticeable proportion of individuals earning '>50K', other groups like 'Black', 'Amer-Indian-Eskimo', and 'Other' have a much smaller proportion earning '>50K' relative to their '<=50K' counterparts.
# - This suggests potential income disparities across different racial groups in the dataset.
```
